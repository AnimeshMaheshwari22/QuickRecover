```python
def calculate_angle(a,b,c):
    a = np.array(a) # First
    b = np.array(b) # Mid
    c = np.array(c) # End
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return angle 
def cal_angle(x1, y1, x2, y2):
    # Calculate the angle between the vertical line passing through A and the line AB
    angle = m.degrees(m.atan2(x2 - x1, y2 - y1))
    
    # Ensure the angle is between 0 and 180
    if angle < 0:
        angle += 180
    
    # Return the angle
    return angle
```
